```python
import warnings
warnings.filterwarnings("ignore")
logger.info("Loading the datasets")
train_df=pd.read_csv('train.csv')
test_df=pd.read_csv('test.csv')
store_df=pd.read_csv('store.csv')
```

```python
logger.info("Checking the null values foe each datasets")
print("Null values present in train_df:")
print(train_df.isnull().sum().sum())
print("Null values present in store_df")
print(store_df.isna().sum().sum())
print("Null valuees present in test_df")
print(test_df.isna().sum().sum())
```

```python
logger.warning('Store_df and test_df has more nunber of null values ')
logger.info("Handling the null values for store_df and test_df")
```

```python
store_df.update(store_df["Promo2SinceWeek"].fillna(value=0,inplace=True))
store_df.update(store_df["Promo2SinceYear"].fillna(value=0,inplace=True))
store_df.update(store_df["PromoInterval"].fillna(value=0,inplace=True))
```

```python
mean_CompetitionDistance=store_df["CompetitionDistance"].mean()
store_df.update(store_df["CompetitionDistance"].fillna(value=mean_CompetitionDistance,inplace=True))
mode_CompetitionOpenSinceMonth=store_df["CompetitionOpenSinceMonth"]=store_df['CompetitionOpenSinceMonth'].mode()[0]
mode_CompetitionOpenSinceYear=store_df["CompetitionOpenSinceYear"]=store_df["CompetitionOpenSinceYear"].mode()[0]
store_df.update(store_df["CompetitionOpenSinceMonth"].fillna(value=mode_CompetitionOpenSinceMonth,inplace=True))
store_df.update(store_df["CompetitionOpenSinceYear"].fillna(value=mode_CompetitionOpenSinceYear,inplace=True))
```

```python
mean_CompetitionDistance=test_df["Open"].mean()
test_df.update(test_df["Open"].fillna(value=mean_CompetitionDistance,inplace=True))
test_df.isna().sum()
```

```python
logger.info('Sucessfully handled the null values for each data set ')
logger.info('Merge the datasets')
train_merged_df=pd.merge(train_df,store_df,on="Store")
test_merged_df=pd.merge(test_df,store_df,on='Store')
```

```python
logger.info('Sucessfully merged the datasets')
```

```python
logger.info("Extract weekdays and weekends from the train and test merged dataset")
train_merged_df['Date']=pd.to_datetime(train_merged_df['Date'])
train_merged_df["WeekDays"]=train_merged_df['Date'].dt.dayofweek+1
train_merged_df['WeekEnds']=train_merged_df['WeekDays'].isin([6,7]).astype(int)
```

```python
test_merged_df['Date']=pd.to_datetime(test_merged_df['Date'])
test_merged_df["WeekDays"]=test_merged_df['Date'].dt.dayofweek+1
test_merged_df['WeekEnds']=test_merged_df['WeekDays'].isin([6,7]).astype(int)
```

```python
train_merged_df
```

```python
logger.info('Extracting begining of the month , mid-month and end of the month ')
# Beginning of the month
train_merged_df.loc[:,"BeginningOfMonth"]=train_merged_df['Date'].dt.is_month_start.astype(int)
# Mid of the month
train_merged_df.loc[:,"MidOfMonth"]=((train_merged_df['Date'].dt.day>10)&(train_merged_df['Date'].dt.day<20)).astype(int)
# End of the month
train_merged_df.loc[:, 'EndOdMonth']=train_merged_df['Date'].dt.is_month_end.astype(int)
```

```python
logger.info('Extracting begining of the month , mid-month and end of the month  for test_merged_df')
# Beginning of the month
test_merged_df.loc[:,"BeginningOfMonth"]=test_merged_df['Date'].dt.is_month_start.astype(int)
# Mid of the month
test_merged_df.loc[:,"MidOfMonth"]=((test_merged_df['Date'].dt.day>10)&(test_merged_df['Date'].dt.day<20)).astype(int)
# End of the month
test_merged_df.loc[:, 'EndOdMonth']=test_merged_df['Date'].dt.is_month_end.astype(int)
```

```python
logger.info('Converting all categorical column value to numerical value')
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
train_merged_df["StateHoliday"]=train_merged_df["StateHoliday"].astype(str)
train_merged_df["PromoInterval"]=train_merged_df["PromoInterval"].astype(str)
train_merged_df['StateHoliday']=label_encoder.fit_transform(train_merged_df['StateHoliday'])
train_merged_df['StoreType']=label_encoder.fit_transform(train_merged_df['StoreType'])
train_merged_df['Assortment']=label_encoder.fit_transform(train_merged_df['Assortment'])
train_merged_df['PromoInterval']=label_encoder.fit_transform(train_merged_df['PromoInterval'])
```

```python
test_merged_df['StateHoliday']=test_merged_df['StateHoliday'].astype(str)
test_merged_df['PromoInterval']=test_merged_df['PromoInterval'].astype(str)
test_merged_df['StateHoliday']=label_encoder.fit_transform(test_merged_df['StateHoliday'])
test_merged_df['StoreType']=label_encoder.fit_transform(test_merged_df['StoreType'])
test_merged_df['Assortment']=label_encoder.fit_transform(test_merged_df['Assortment'])
test_merged_df['PromoInsterval']=label_encoder.fit_transform(test_merged_df['PromoInterval'])
```

```python
columns_to_scaled=['Sales','Customers','CompetitionDistance']
scaler=StandardScaler()
scaler.fit(train_merged_df[columns_to_scaled])
train_merged_df[columns_to_scaled]=scaler.transform(train_merged_df[columns_to_scaled])
train_merged_df
```

```python
column_to_scaled_test=['CompetitionDistance']
scaler.fit(test_merged_df[column_to_scaled_test])
test_merged_df[column_to_scaled_test]=scaler.transform(test_merged_df[column_to_scaled_test])
test_merged_df
```
